# Bdnews24

```python
import os
import requests
from bs4 import BeautifulSoup
from datetime import datetime
from datetime import date
import re
```

```python
def txtret(li):
    try:
        page = requests.get(li)
    except:
        return ""
    soup = BeautifulSoup(page.content, 'html.parser')
    txt=soup.select('p')
    return txt
def filename():
    now = datetime.now()
    today = date.today()
    current_time = now.strftime("%H_%M_%S")
    d1 = today.strftime("%d_%m_%Y")
    file_name=str(d1+"_"+current_time)
    return file_name
def soupcalled(lin):
    page = requests.get(lin)
    soup = BeautifulSoup(page.content, 'html.parser')   
    return soup
```

```python
root=r"C:\Users\Administrator\Desktop\my research(Shakib)\tokenizer\prothom alo data collect"
root2=root+"\raw"
bdnews_already_crawled=open(os.path.join(root,"bdnews24.txt"),"r").read().split("\n")
bdnews_already_crawled
```

```python
soup=soupcalled("https://bangla.bdnews24.com/")
bdnews=[]
for x in soup.select('a'):
    link_org=x.get('href')
    if link_org==None:
        continue
    elif link_org.find('collection')!=-1:
        soup2=soupcalled(link_org)
        for xx in soup2.select('a'):
            link_org2=xx.get('href')
            if link_org2!=None and link_org2[8:27]=="bangla.bdnews24.com" and link_org2.find("collection")==-1:
                bdnews.append(link_org2)
                print(f"collection: {link_org2}")
            
    elif link_org[8:27]=="bangla.bdnews24.com":
        print(link_org)
        bdnews.append(link_org2)
        
```

```python
data2=[]
for x in bdnews:
    link_org=x
    link=link_org[8:]
    sp_link=link.split("/")
    if (len(sp_link)==3 or len(sp_link)==4) and sp_link[len(sp_link)-1] not in bdnews_already_crawled:
        print("Crawling started-----")
        bdnews_already_crawled.append(link.split("/")[len(sp_link)-1])
    else:
        print("Already crawled!!!")
        continue
    if len(sp_link)==3 or len(sp_link)==4:
        if link.find("play.google.com")!=-1:
            continue
        print(sp_link)
        res=str(txtret(link_org))
        res=re.sub(r'<p>','',res)
        res=re.sub(r'</p>','',res)
        res=re.sub(r'\[','',res)
        res=re.sub(r'\]','',res)
        res=re.sub(r'[a-zA-z0-9]|<|=|"|>|©','',res)
        data2.append(res)
```

```python
if len(data2)!=0:
    f2=open(os.path.join(root2,filename()+"bdnews24"),"w",encoding="utf-8")
    data2="।".join(data2)
    f2.write(data2)
    f2.close()
```

```python
f=open(os.path.join(root,"bdnews24.txt"),"w")
bdnews_already_crawled="\n".join(bdnews_already_crawled)
f.write(bdnews_already_crawled)
f.close()
```
